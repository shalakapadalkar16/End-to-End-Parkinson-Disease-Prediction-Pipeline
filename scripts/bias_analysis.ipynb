```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Load the dataset
df = pd.read_csv('/home/yutika/Downloads/final_cleaned.csv') 
```

```python
# Plot distribution of Age (ENROLL_AGE)
plt.figure(figsize=(10, 6))
df['ENROLL_AGE'].plot(kind='hist', bins=20, color='skyblue', edgecolor='black')
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# Plot distribution of Gender (SEX)
plt.figure(figsize=(10, 6))
df['SEX'].value_counts().plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title("Distribution of Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

# Plot distribution of Cohort (COHORT)
plt.figure(figsize=(10, 6))
df['COHORT'].value_counts().plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title("Distribution of Cohort")
plt.xlabel("Cohort")
plt.ylabel("Count")
plt.show()
```
